```python
#6.7.4
import numpy as np
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt

def fdjac(f, x0, y0):

    delta = np.sqrt(np.finfo(float).eps) 
    m, n = len(y0), len(x0)
    J = np.zeros((m, n))
    I = np.eye(n)
    for j in range(n):
        J[:, j] = (f(x0 + delta * I[:, j]) - y0) / delta
    return J

def levenberg(f, x1, tol=1e-12):
    ftol = tol
    xtol = tol
    maxiter = 40

    n = len(x1)
    x = np.zeros((maxiter+1, n))
    x[0] = x1
    fk = f(x1)
    k = 0
    s = 10.0
    Ak = fdjac(f, x[0], fk)  
    jac_is_new = True

    lam = 10
    while (np.linalg.norm(s) > xtol) and (np.linalg.norm(fk) > ftol) and (k < maxiter):
        
        B = Ak.T @ Ak + lam * np.eye(n)
        z = Ak.T @ fk
        s = -np.linalg.lstsq(B, z)[0]

        xnew = x[k] + s
        fnew = f(xnew)

        if np.linalg.norm(fnew) < np.linalg.norm(fk):  # accept
            y = fnew - fk
            x[k + 1] = xnew
            fk = fnew
            k = k + 1

            lam = lam / 10  
            Ak = Ak + np.outer(y - Ak @ s, s / np.dot(s, s))
            jac_is_new = False
        else:  
            lam = lam * 4
            if not jac_is_new:
                Ak = fdjac(f, x[k], fk)
                jac_is_new = True

    if np.linalg.norm(fk) > 1e-3:
        warnings.warn("Iteration did not find a root.")
    return x[:k+1]
    
def am2(du_dt, tspan, u0, n):

    a, b = tspan
    h = (b - a) / n
    t = np.linspace(a, b, n + 1)

    u = np.zeros((n+1, len(np.atleast_1d(u0))))
    u[0] = u0

    
    for i in range(n):
        known = u[i] + h / 2 * du_dt(t[i], u[i])
        F = lambda z: z - h / 2 * du_dt(t[i+1], z) - known
        unew = levenberg(F, known)
        u[i+1] = unew[-1]

    return t, u.T

du_dt = lambda t, u: -2 * t * u
tspan = [0.0, 2.0]
n = [10 * 2**d for d in range(1, 11)]
u0 = [2.0]
error = np.zeros(len(n)) 
u_exact = lambda t: 2 * np.exp(-t**2)

fig, ax = plt.subplots()
for i in range(0, 10):
    t, u = ab4(du_dt, tspan, u0, n[i])
    error[i] = np.linalg.norm(np.abs(u[0] - u_exact(t)), np.inf)
        
ax.loglog(n, error, label = "error")
ref_error_2th = error[0] * (np.array(n) / n[0])**(-2)
ax.loglog(n, ref_error_2th, 'k--', label="2th order ref (slope=-2)")
plt.grid()
plt.legend()
plt.show()
```
