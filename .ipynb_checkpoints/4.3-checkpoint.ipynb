```python
import numpy as np

def f(x):
    return np.exp(x + 1) - 2 - x
def df(x):
    return np.exp(x + 1) - 1

x_current = 0.0 
max_iterations = 30

for k in range(max_iterations):
    
    print(f"x = {x_current:.15f}")
    
    f_val = f(x_current)
    df_val = df(x_current)
    
    if abs(df_val) < 1e-14:
        print("df(x)->0，STOP")
        break
    
    x_next = x_current - f_val / df_val
    
    if abs(x_next - x_current) < 1e-15:
        print("已收敛。")
        x_current = x_next
        break
        
    x_current = x_next


print(f"x = {x_current:.15f}")
```
