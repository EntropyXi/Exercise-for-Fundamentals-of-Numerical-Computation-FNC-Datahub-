```python
#4.4.1
import numpy as np
import matplotlib.pyplot as plt
import scipy

def f_a(x):
    return np.exp(-x) - x**2
def secant(f, x0, x1, maxiter):
    #割线法
    x = np.zeros(maxiter)
    x[0] = x0
    x[1] = x1
    epsm = np.finfo(float).eps
    y = 100 * epsm
    k = 1
    
    while np.abs(f(x[k])) > y and k < maxiter - 1:
        #迭代循环
        x[k+1] = x[k] - (f(x[k]) * (x[k] - x[k-1])) / (f(x[k]) - f(x[k-1]))
        k += 1
        #检查分母
        denom = f(x[k]) - f(x[k-1])
        if abs(denom) < epsm:  
            print("error")
            break  # 若分母过小则提前跳出循环
    
    eps = np.abs(x - x[k])
    print(eps)
    return x[k]

xx = np.linspace(-2, 2, 500)
r1 = scipy.optimize.root_scalar(f_a, bracket = [0.5, 1.0]).root
r2 = secant(f_a, 0.5, 1, 30)

print(f"scipy求根：{r1}")
print(f"secant求根{r2}")

fig, ax = plt.subplots()
ax.plot(xx, f_a(xx))
ax.grid()
```

```python
#4.4.2
import numpy as np
import matplotlib.pyplot as plt

def f_b(x):
    return np.tan(x) - 2*x
def secant(f, x0, x1, maxiter):
    #割线法
    x = np.zeros(maxiter)
    x[0] = x0
    x[1] = x1
    epsm = np.finfo(float).eps
    y = 100 * epsm
    k = 1
    
    while np.abs(f(x[k])) > y and k < maxiter - 1:
        
        #检查分母
        denom = f(x[k]) - f(x[k-1])
        if abs(denom) < epsm:  
            print("error")
            break  # 若分母过小则提前跳出循环
        #迭代循环
        x[k+1] = x[k] - (f(x[k]) * (x[k] - x[k-1])) / (f(x[k]) - f(x[k-1]))
        k += 1
        
    eps = np.abs(x - x[k])
    print(eps)
    return x[k]
        
xx = np.linspace(-0.2, 1.4, 500)
r1 = secant(f_b, -0.1, 0.1, 30)
r2 = secant(f_b, 1.0, 1.2, 30)

print(f"secant求左根{r1}")
print(f"secant求右根{r2}")

fig, ax = plt.subplots()
ax.plot(xx, f_b(xx))
ax.grid()
```

```python
#4.4.9
import numpy as np
import matplotlib.pyplot as plt

f = lambda x: np.exp(-x) - x**2
def iqi(f, x0, x1, x2, maxiter = 50):
    #逆二次插值
    x = np.zeros(maxiter)
    x[0] = x0
    x[1] = x1
    x[2] = x2
    epsm = np.finfo(float).eps
    y = 100 * epsm
    k = 2

    while np.abs(f(x[k])) > y and k < maxiter - 1:
        x_pts = x[k-2:k+1]
        y_pts = f(x_pts)
        q = np.poly1d(np.polyfit(y_pts, x_pts, 2))
        iq = np.poly1d(np.polyfit(y_pts, x_pts, 2))
        x[k+1] = iq(0)

    k += 1
    
    return x[k]

r = iqi(f, 0.5, 1, 0.75)
print(r)
    
    
    
```
