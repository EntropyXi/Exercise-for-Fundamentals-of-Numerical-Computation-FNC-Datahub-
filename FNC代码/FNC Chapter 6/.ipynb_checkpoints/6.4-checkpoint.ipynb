```python
import numpy as np
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt
from prettytable import PrettyTable

def ex_ie(du_dt, tspan, u0, n):
    
    a, b = tspan
    h = (b - a) / n
    t = np.linspace(a, b, n + 1)

    u = np.zeros((n+1, len(np.atleast_1d(u0))))
    u[0] = u0

    # 迭代
    for i in range(n):
        k1 = h * du_dt(t[i], u[i])
        k2 = h * du_dt(t[i] + (1/2)*h, u[i] + k1/2)
        u[i+1] = u[i] + (1/2)*k1 + (1/2)*k2

    return t, u.T

du_dt = lambda t, u: -2 * t * u
u0 = 2
tspan = [0.0, 2.0]
n = np.arange(30, 301, 30)
sol = solve_ivp(du_dt, tspan, [u0], dense_output=True)
u_ref = sol.sol

err = {"ex_ie" : []}
results = PrettyTable(["n", "ex_ie error"])
for i in range(len(n)):
    t, u = ex_ie(du_dt, tspan, u0, n[i])
    err["ex_ie"].append( abs(u_ref(4)[0] - u[0][-1]) )
    results.add_row([n[i], err["ex_ie"][-1]])
print(results)

plt.loglog(2 * n, err["ex_ie"], "-o", label="ex_ie")
plt.plot(2 * n, 0.5 * err["ex_ie"][-1] * (n / n[-1])**(-2), "--", label="2nd order")
```
