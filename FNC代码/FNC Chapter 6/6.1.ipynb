```python
#6.1.2(a)
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp

f = lambda t, u: 3 * u
tspan = [0.0, 1.0]
u0 = np.array([1.0])
sol = solve_ivp(f, tspan, u0)

plt.plot(sol.t, sol.y[0], 'o-')
```

```python
#6.1.5
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp

k = 1
S = 1
M = 0.25
f = lambda t, x: k * (S - x) * (x - M)
tspan = [0.0, 3.0]
x0 = np.array([[0.9*M], [1.1*M], [1.5*M], [0.9*S], [1.1*S], [3.0*S]])
for i in range(len(x0)):
    sol = solve_ivp(f, tspan, x0[i])
    plt.plot(sol.t, sol.y[0], 'o-')
```

```python
#6.1.6(a)(b)
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp

f = lambda t, u: u * np.cos(u) + np.cos(4*t)
tspan = [0, 10]
u0 = np.array([[-0.058], [0.0]])

for i in range(len(u0)):
    sol = solve_ivp(f, tspan, u0[i])
    plt.plot(sol.t, sol.y[0], 'p-')
```

```python
#6.1.7
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp

k = np.log(2) / 8
f1 = lambda t, x: -k * x + 16
f2 = lambda t, x: -k * x
tspan1 = [0.0, 0.5]
tspan2 = [0.5, 12.0]
x0 = np.array([0.0])

sol1 = solve_ivp(f1, tspan1, x0)
sol2 = solve_ivp(f2, tspan2, x0)

plt.plot(sol1.t, sol1.y[0], 'o-')
plt.plot(sol2.t, sol2.y[0], 'o-')
```
