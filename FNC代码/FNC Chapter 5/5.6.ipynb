```python
#5.6.1
import scipy as sp
import numpy as np
import matplotlib.pyplot as plt

def trapezoid(f, a, b, n):
    h = (b - a) / n
    t = np.linspace(a, b, n + 1)
    y = f(t)
    T = h * (np.sum(y[1:-1]) + 0.5 * (y[0] + y[-1]))
    
    return T, t, y

f = lambda x: x * np.log(1+x)
a = 0
b = 1
I = sp.integrate.quad(f, a, b)[0]
N = np.array([10 * 2**k for k in range(1, 11)])
Tf = np.array([trapezoid(f, a, b, n)[0] for n in N])
errors = abs(I - Tf)

ax, fig = plt.subplots()

plt.loglog(N, errors)



```
