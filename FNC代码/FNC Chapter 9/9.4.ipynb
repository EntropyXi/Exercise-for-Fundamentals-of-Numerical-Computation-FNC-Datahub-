```python
#9.4.5 and 9.4.6
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

def legpoly(x, n):
    V = np.zeros((len(x), n + 1))
    V[:, 0] = np.ones(len(x))
    V[:, 1] = x.T
    for k in range(2, n + 1):
        V[:, k] = ( (2*k - 1)/k ) * x * V[:, k-1] - ( (k-1)/k ) * V[:, k-2]

    return V

x = np.linspace(-1, 1, 1600)
N = np.array([i for i in range(1, 17)])
kappa = []
for n in N:
    A = legpoly(x, n)
    _, s, _ = np.linalg.svd(A)
    kappa.append(s[0] / s[-1])

kappa_n = {
    'n': N,
    'kappa': kappa
}
df = pd.DataFrame(kappa_n)
print(df)
```

```python
#9.4.7
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

def chebpoly(x, n):
    T = np.zeros((len(x), n + 1))

    for k in range(n + 1):
        Tk = np.cos(k * np.arccos(x))
        T[:, k] = Tk.T

    return T

x_plot = np.linspace(-1, 1, 200)
n_degree = 3

T_matrix = chebpoly(x_plot, n_degree)

plt.figure(figsize=(10, 6))
plt.plot(x_plot, T_matrix, linewidth=2)

plt.title('Chebyshev Polynomials $T_0, T_1, T_2, T_3$', fontsize=14)
plt.xlabel('x')
plt.ylabel('$T_k(x)$')
plt.axhline(0, color='black', linewidth=0.5, linestyle='--')
plt.axvline(0, color='black', linewidth=0.5, linestyle='--')
plt.grid(True, alpha=0.3) # 添加网格，透明度 0.3
plt.legend(['$T_0(x)=1$', '$T_1(x)=x$', '$T_2(x)=2x^2-1$', '$T_3(x)=4x^3-3x$'])
plt.ylim([-1.2, 1.2])

plt.show()
```
