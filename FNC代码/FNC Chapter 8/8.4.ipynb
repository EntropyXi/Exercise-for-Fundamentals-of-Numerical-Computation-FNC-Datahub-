```python
#8.4
import numpy as np
import scipy as sp

def arnoldi(A, u, m):
    n = A.shape[0]
    Q = np.zeros([n, m + 1])
    H = np.zeros([m + 1, m])
    Q[:, 0] = u / np.linalg.norm(u)

    for i in range(m):
        v =  A @ Q[:, i]

        for j in range(i + 1):
            H[j, i] = Q[:, j] @ v
            v -= H[j, i] * Q[:, i]

        H[i + 1, i] = np.linalg.norm(v)
        Q[:, i + 1] = v / H[i + 1, i]

    return Q, H
```

```python
#8.2.7
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt

def arnoldi(A, u, m):
    n = A.shape[0]
    Q = np.zeros([n, m + 1])
    H = np.zeros([m + 1, m])
    Q[:, 0] = u / np.linalg.norm(u)

    for i in range(m):
        v =  A @ Q[:, i]

        for j in range(i + 1):
            H[j, i] = Q[:, j] @ v
            v -= H[j, i] * Q[:, j]

        H[i + 1, i] = np.linalg.norm(v)
        Q[:, i + 1] = v / H[i + 1, i]

    return Q, H

ev = 10 + np.arange(1, 101)
maxA = 110
A = np.triu(np.random.rand(100, 100), 1) + np.diag(ev)
b = np.random.rand(100)
m = np.arange(1, 41)
err = []
for i in m:
    Q, H = arnoldi(A, b, i)
    ev_Hms = np.linalg.eig(H[:-1, :])[0]
    maxH = ev_Hms[np.argmax(np.abs(ev_Hms))]
    err.append(abs(maxH - maxA))

plt.subplots()
plt.semilogy(m, err, '-o')
plt.grid(True)
plt.show()

```
