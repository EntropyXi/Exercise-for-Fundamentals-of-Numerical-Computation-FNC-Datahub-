```python
#5.2.1(a)
import numpy as np
import matplotlib.pyplot as plt

def plinterp(t, y):
    n = len(t) - 1
    
    def hatfunc(t, k): #t为所有的x，k为你想要调用的x的索引
        def evaluate_hat(x):
            H = np.zeros(np.array(x).shape)

            for j, xj in enumerate(x):
                if (k > 0) and (xj >= t[k-1]) and (xj <= t[k]):
                    H[j] = (xj - t[k-1]) / (t[k] - t[k-1])
                elif (k < n) and (xj >= t[k]) and (xj <= t[k+1]):
                    H[j] = (t[k+1] - xj) / (t[k+1] - t[k])

            return H
        return evaluate_hat

    H = [hatfunc(t, k) for k in range(n+1)]
    def evaluate_plin(x):
        f = 0
        for k in range(n+1):
            f += y[k] * H[k](x)
        return f
    return evaluate_plin

n_values = np.array([10, 20, 40, 80, 160, 320, 640])
x = np.linspace(0, 4, 1600)
f_a = lambda x: np.cos(np.pi * x**2)
f_eval = f_a(x)

E = []
for n in n_values:
    t = np.linspace(0, 4, n+1)
    y = f_a(t)
    
    p_a = plinterp(t, y)
    p_eval = p_a(x)

    max_e = np.max(np.abs(f_eval - p_eval))

    E.append(max_e)

fig, ax = plt.subplots()
ax.loglog(n_values, E, 'o-', label='E(n)')

C = E[0] * (n_values[0]**2)
E_ref = C * (n_values**2)
ax.loglog(n_values, E_ref, label=f'E = C * n^-2 (参考线, C≈{C:.2f})')
```

```python
#5.2.1(b)
import numpy as np
import matplotlib.pyplot as plt

def plinterp(t, y):
    #分段线性插值函数 返回分段线性插值函数值
    n = len(t) - 1

    def hatfunc(t, k):
        #返回一段hat函数
        def evaluate_hat(x):
            H = np.zeros(np.array(x).shape)
            #确定需求的一段H
            for j, xj in enumerate(x):
                if (k > 0) and (xj >= t[k-1]) and (xj <= t[k]):
                    H[j] = (xj - t[k-1]) / (t[k] - t[k-1])
                elif (xj >= t[k]) and (xj <= t[k+1]):
                    H[j] = (t[k+1] - xj) / (t[k+1] - t[k])
            return H
        return evaluate_hat

    H = [hatfunc(t, k) for k in range(n+1)] #求出每一段hat 并汇总到一个列表里
    def evaluate_plin(x):
        #总函数 求出y*phi(x)之和 构造拉格朗日插值函数
        f = 0
        for k in range(n+1):
            f += y[k] * H[k](x)
        return f

    return evaluate_plin

f_b = lambda x: np.log(x)
n_values = np.array([10, 20, 40, 80, 160, 320, 640])
test_points = np.linspace(1, 20, 1600)
truth_at_test = f_a(test_point)

E = []
for n in n_values:
    nodes = np.linspace(1, 20, n+1)
    y_nodes = f_b(nodes)
    #构造插值函数
    my_interpolant_func = plinterp(nodes, y_nodes) 
    #计算插值在 1600 个测试点上的值
    approx_at_test = my_interpolant_func(test_points)
    #计算最大误差
    current_error = np.max(np.abs(truth_at_test - approx_at_test))
    E.append(current_error)

fig, ax = plt.subplots()
# 4. 画 Log-Log 图
plt.loglog(n_values, E, label='My Error', marker='o')

#画参考线 E = C * n^-2
#C选一个让线靠近数据的数，比如 errors[0] * (n_values[0]**2)
C = E[0] * (n_values[0]**2) 
ref_line = [C * (float(n)**-2) for n in n_values]
plt.loglog(n_values, ref_line, label='Ref n^-2', linestyle='--')

plt.legend()
plt.show()
```
