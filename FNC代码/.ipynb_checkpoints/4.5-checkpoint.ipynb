```python
import numpy as np

def f(x_vec):
    u = x_vec[0]
    v = x_vec[1]

    f1 = u * np.log(u) + v * np.log(v) + 0.3
    f2 = u**4 + v**2 - 1

    return np.array([f1, f2])
    
def jac(x_vec):
    u = x_vec[0]
    v = x_vec[1]
    
    df1_u = np.log(u) + 1
    df1_v = np.log(v) + 1
    df2_u = 4 * u**3
    df2_v = 2 * v
    
    return np.array([
        [df1_u, df1_v],
        [df2_u, df2_v]
    ])

def newton(f, jac, x1, maxiter):
    #初始化
    x = np.zeros((maxiter, len(x1)))
    x[0] = x1
    y, J = f(x1), jac(x1)
    dx = 10.0
    k = 0

    #迭代
    while np.linalg.norm(y) > 1e-12 and k < maxiter - 1:
        dx = - np.linalg.lstsq(J, y)[0]

        x[k+1] = x[k] + dx

        k = k + 1
        y, J = f(x[k]), jac(x[k])
    
    return x[k]

msg = newton(f, jac, np.array([1.0, 0.1]), 30)
print(msg)
```

```python
import numpy as np

def f(x_vec):
    #定义函数f
    t1 = x_vec[0]
    t2 = x_vec[1]
    #求偏导
    f1 = 10*np.cos(t1) - 8*np.cos(t2) - 5
    f2 = 6*np.sin(t1) - 12*np.sin(t2) - 3
    
    return np.array([f1, f2])

def jac(x_vec):
    #定义f的雅可比矩阵
    t1 = x_vec[0]
    t2 = x_vec[1]
    #求偏导
    df1_t1 = -10*np.sin(t1)
    df1_t2 = 8*np.sin(t2)
    df2_t1 = 6*np.cos(t1)
    df2_t2 = -12*np.cos(t2)

    return np.array([
        [df1_t1, df1_t2],
        [df2_t1, df2_t2]
    ])

def newton(f, jac, x1, maxiter = 30):
    x = np.zeros((maxiter, 2))
    x[0] = x1
    
    dx = 10.0
    k = 0
    y = f(x[k])
    J = jac(x[k])
    #迭代
    while np.linalg.norm(y) > 1e-12 and k < maxiter - 1:
        dx = -np.linalg.lstsq(J, y)[0]
        x[k+1] = x[k] + dx
        
        k += 1
        y, J = f(x[k]), jac(x[k])
    
    return x[k]

def find_all_intersections(f, jac, newton, n_points=15):
    search_place = np.linspace(0, 2*np.pi, n_points)

    unique_xy = set() #定义一个集合用于后面的去重（利用集合元素不相等的特性）
    tolerance = 1e-12 #精度
    rounding_decimals = 8 #定义后面四舍五入去重时保留的小数点位数
    
    for t1 in search_place:
        for t2 in search_place:
            guess_x = np.array([t1, t2])
            root = newton(f, jac, guess_x)
            
            if np.linalg.norm(f(root)) < tolerance:
                
                # 5. 坐标转换
                '''
                newton函数所求出来的根只是时间
                所以要把时间转化为xy坐标
                '''
                x_coord = -5 + 10 * np.cos(root[0]) # t1
                y_coord = 6 * np.sin(root[0])   # t1
                
                # 6. 去重
                '''
                列表是可变的不能塞进去集合
                而把坐标化入元组（不可变）就能塞入集合了
                '''
                rounded_xy = (
                    round(x_coord, rounding_decimals), 
                    round(y_coord, rounding_decimals)
                )
                unique_xy.add(rounded_xy)
                
    return unique_xy

all_points = find_all_intersections(f, jac, newton, n_points=20)
if all_points:         #如果all_points不为空
    for i, point in enumerate(all_points):  #enumerate()生成一个带有索引的可迭代对象
                                            #在遍历all_point的同时 提供一个计数器i
        print(point)
        
else:
    print("None")    
```

```python
import numpy as np

def f(x_vector):
    x = x_vector[0]
    y = x_vector[1]
    z = x_vector[2]
    lam = x_vector[3]

    f1 = x - lam*x / 25 - 5
    f2 = y - lam*y / 16 - 4
    f3 = z - lam*z / 9 - 3
    f4 = x**2/25 + y**2/16 + z**2/9 - 1

    return np.array([f1, f2, f3, f4])

def jac(x_vector):
    x = x_vector[0]
    y = x_vector[1]
    z = x_vector[2]
    lam = x_vector[3]

    return np.array([
        [-lam / 25 + 1, 0, 0, -x / 25],
        [0, -lam / 16 + 1, 0, -y / 16],
        [0, 0, -lam / 9 + 1, -z / 9],
        [2*x / 25, 2*y / 16, 2*z / 9, 0]
    ])

def newton(f, jac, x1, maxiter = 30):
    x = np.zeros((maxiter, 4))
    x[0] = x1

    k = 0
    dx = 10.0
    y, J = f(x[k]), jac(x[k])

    while np.linalg.norm(y) > 1e-12 and np.linalg.norm(dx) > 1e-12 and k < maxiter - 1:
        dx = -np.linalg.lstsq(J, y)[0]
        x[k+1] = x[k] + dx
        
        k += 1
        y, J = f(x[k]), jac(x[k])

    if k == maxiter and np.linalg.norm(y) > 1e-12:
        print(f"已迭代{k}次，未收敛至根")

    return x[k]

guess1 = np.array([5.0, 4.0, 3.0, 0.0])
guess2 = np.array([-5.0, -4.0, -3.0, 30.0])
guesses = [guess1, guess2]
solutions = []

for i, guess in enumerate(guesses):
    root = newton(f, jac, guess)
    
    if np.linalg.norm(f(root)) < 1e-12:
        solutions.append(root)
    else:
        print(f"猜测 {i+1} 未能收敛到有效解。")

for sol in solutions:
    # 提取前三个元素
    print(f"({sol[0]:.6f}, {sol[1]:.6f}, {sol[2]:.6f})")
```

```python
import numpy as np
import matplotlib.pyplot as plt

def f(v):
    a = v[0]
    b = v[1]
    r = v[2]

    f1 = (a - 5)**2 + (b - 0)**2 - r**2
    f2 = (a - 1)**2 + (b + 3)**2 - r**2
    f3 = (a - 4)**2 + (b - 2)**2 - r**2

    return np.array([f1, f2, f3])

def jac(v):
    a = v[0]
    b = v[1]
    r = v[2]

    return np.array([
        [2 * (a - 5), 2 * b, -2 * r],
        [2 * (a - 1), 2 * (b + 3), -2 * r],
        [2 * (a - 4), 2 * (b - 2), -2 * r]
    ])

def newtonsys(f, jac, x1, maxiter = 50):
    x = np.zeros((maxiter, 3))
    x[0] = x1

    k = 0
    dx = 10.0
    y, J = f(x[k]), jac(x[k])

    while np.linalg.norm(y) > 1e-12 and np.linalg.norm(dx) > 1e-12 and k < maxiter - 1:
        dx = -np.linalg.lstsq(J, y)[0]
        x[k+1] = x[k] + dx
        
        k += 1
        y, J = f(x[k]), jac(x[k])

    if k == maxiter and np.linalg.norm(y) > 1e-12:
        print(f"已迭代{k}次，未收敛至根")

    return x[k]

x1 = np.array([3, 3, 5])
root = newtonsys(f, jac, x1)
print(root)

a = root[0]
b = root[1]
r = root[2]

t = np.linspace(0, 2 * np.pi, 200)
# 2. 用参数方程创建 (x, y) 坐标
xx = a + r * np.cos(t)
yy = b + r * np.sin(t)

fig, ax = plt.subplots()
plt.plot(xx, yy)
plt.grid()
plt.show()
```
