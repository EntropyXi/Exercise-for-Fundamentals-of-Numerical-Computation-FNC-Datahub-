```python
#4.2.2(a)
import numpy as np
import matplotlib.pyplot as plt

def g_a(x):
    return 1.0 + x - (1.0/9.0) * x**2
def logek(k, sigma, C):
    return k * np.log(sigma) + np.log(C)

p_a = 3
sigma_a = np.abs(1 - 2.0/9.0 * p_a) #sigma=|g'(p)|
x = np.zeros(26)
x[0] = 1

kk = np.linspace(1, 26, 500)

fig, ax = plt.subplots()
for k in range(25):
    x[k+1] = g_a(x[k])

epsilon = np.abs(x - p_a)
#绘制实际线
k_values_for_plot = np.arange(1, 27)
ax.plot(k_values_for_plot, np.log(epsilon))

# 我们使用 e_1 估计: C ≈ e_1 / sigma^1
C_estimate = epsilon[0] / sigma_a
# 计算理论线
theoretical_log_error = logek(kk, sigma_a, C_estimate)
#绘图
ax.plot(kk, theoretical_log_error)
ax.grid()
ax.set_xlabel("$iteration number$"), ax.set_ylabel("error$")
```
