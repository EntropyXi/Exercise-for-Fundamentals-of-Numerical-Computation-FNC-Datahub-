```python
#
import numpy as np
import matplotlib.pyplot as plt
def spinterp_moment(t, y):
    """
     (Method of Moments) 
    t:  ()
    y: 
    :  evaluate 
    """
    n = len(t) - 1
    h = np.diff(t)  #  h[i] = t[i+1] - t[i]

    #
    A = np.zeros((n + 1, n + 1))
    b = np.zeros(n + 1)
    
    # : M_0 = 0, M_n = 0
    # 
    A[0, 0] = 1
    A[n, n] = 1
    # b[0]  b[n]  0
    
    #  (i = 1  n-1)
    for i in range(1, n):
        #  mu  lambda
        mu = h[i-1] / (h[i-1] + h[i])
        lam = h[i] / (h[i-1] + h[i])
        
        #  A
        A[i, i-1] = mu
        A[i, i]   = 2
        A[i, i+1] = lam
        
        #  d_i 
        div_diff = (y[i+1] - y[i])/h[i] - (y[i] - y[i-1])/h[i-1]
        b[i] = 6 * div_diff / (h[i-1] + h[i])
        
    # 
    M = np.linalg.solve(A, b)
    
    #  M
    def evaluate(x_in):
        #  x_in 
        x_arr = np.atleast_1d(x_in)
        y_out = np.zeros_like(x_arr, dtype=float)
        
        #  x 
        for k, val in enumerate(x_arr):
            #  val  idx t[idx] <= val <= t[idx+1]
            # searchsorted  1 
            idx = np.searchsorted(t, val) - 1
            
            #  x 
            if idx < 0: idx = 0
            if idx >= n: idx = n - 1
            
            # 
            hi = h[idx]        # 
            xi = t[idx]        #  x
            xi1 = t[idx+1]     #  x
            yi = y[idx]        #  y
            yi1 = y[idx+1]     #  y
            Mi = M[idx]        # 
            Mi1 = M[idx+1]     # 
            
            #  ( M )
            # Term 1 & 2: 
            term1 = Mi * (xi1 - val)**3 / (6 * hi)
            term2 = Mi1 * (val - xi)**3 / (6 * hi)
            # Term 3 & 4: 
            term3 = (yi - Mi * hi**2 / 6) * (xi1 - val) / hi
            term4 = (yi1 - Mi1 * hi**2 / 6) * (val - xi) / hi
            
            y_out[k] = term1 + term2 + term3 + term4
            
        # 
        if np.isscalar(x_in):
            return y_out[0]
        return y_out
    
    return evaluate

f_true = lambda x: np.cos(np.pi * x**2)

#Gemini()
#  n  (10, 20, ..., 640)
n_values = np.array([10, 20, 40, 80, 160, 320, 640])
errors = []

print(f"{'n':<10} | {'Max Error':<20} | {'Rate (Slope)':<10}")
print("-" * 50)

for i, n in enumerate(n_values):
    # 1. 
    nodes = np.linspace(0, 4, n + 1)
    y_nodes = f_true(nodes)
    
    # 2. 
    spline_func = spinterp_moment(nodes, y_nodes)
    
    # 3.  ()
    test_points = np.linspace(0, 4, 1600)
    y_test_true = f_true(test_points)
    y_test_spline = spline_func(test_points)
    
    # 
    max_err = np.max(np.abs(y_test_true - y_test_spline))
    errors.append(max_err)
    
    #  (Slope)
    rate = 0.0
    if i > 0:
        # Slope = log(E_curr / E_prev) / log(n_curr / n_prev)
        rate = np.log(max_err / errors[i-1]) / np.log(n / n_values[i-1])
        
    print(f"{n:<10} | {max_err:<20.5e} | {rate:<10.2f}")

# --- 3.  Log-Log  ---
plt.figure(figsize=(10, 6))

#  ()
plt.loglog(n_values, errors, 'bo-', linewidth=2, label='Actual Spline Error')

# : O(n^-4) ()
# 
C = errors[0] * (n_values[0]**2)
ref_line = [C * (float(n)**-4) for n in n_values]
plt.loglog(n_values, ref_line, label='Ref n^-4', linestyle='--')

plt.xlabel('Number of Intervals (n)')
plt.ylabel('Max Error $E(n)$')
plt.title('Error Convergence Analysis (Log-Log Plot)')
plt.grid(True, which="both", ls="-", alpha=0.5)
plt.legend()

# 
plt.text(n_values[3], errors[3], f"Slope $\\approx$ {rate:.1f}", verticalalignment='bottom')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def spinterp(t, y):
    #
    n = len(t) - 1
    #
    h = [t[i + 1] - t[i] for i in range(n)] #h
    Z = np.zeros([n, n]) #
    H = np.diag(h) #Hh
    J = np.eye(n) + np.diag(-np.ones(n - 1), 1) #
    I = np.eye(n) #
    E = I[: n - 1, :]

    #
    AL = np.hstack([I, Z, Z, Z])
    vL = y[:-1] #

    #
    AR = np.hstack([I, H, H**2, H**3])
    vR = y[1:] #

    #
    A1 = E @ np.hstack([Z, J, 2*H, 3 * H**2])
    v1 = np.zeros(n - 1)

    #
    A2 = E @ np.hstack([Z, Z, J, 3 * H])
    v2 = np.zeros(n - 1)

    natL = np.hstack([np.zeros(2 * n), np.hstack([1, np.zeros(n - 1)]), np.zeros(n)])

    h_end = h[-1] # 
    # c: [0...0, 1]
    c_part = np.hstack([np.zeros(n - 1), 1])
    # d: [0...0, 3*h_end]
    d_part = np.hstack([np.zeros(n - 1), 3 * h_end])
    
    natR = np.hstack([np.zeros(2 * n), c_part, d_part])

    A = np.vstack([AL, AR, A1, A2, natL, natR])
    v = np.hstack([vL, vR, v1, v2, 0, 0]) #  0
    z = np.linalg.solve(A, v)

    #abcd
    rows = np.arange(n)
    a = z[rows]
    b = z[n + rows]
    c = z[2 * n + rows]
    d = z[3 * n + rows]
    
    #abcd
    S = [np.poly1d([d[k], c[k], b[k], a[k]]) for k in range(n)]

    #
    def evaluate(x):
        f = np.zeros(x.shape)
        
        for k in range(n):
            index = (x >= t[k]) & (x <= t[k + 1])
            f[index] = S[k](x[index] - t[k])
        
        return f

    return evaluate #spinterp() 

n_values = np.array([10, 20, 40, 80, 160, 320, 640])
f_a = lambda x: np.cos(np.pi * x**2)
test_points = np.linspace(0, 4, 1600)
truth_at_test = f_a(test_points)

errors = []
for n in n_values:
    nodes = np.linspace(0, 4, n+1)
    y_nodes = f_a(nodes)
    #
    f_cubic = spinterp(nodes, y_nodes)
    #
    cubic_evaluate = f_cubic(test_points)
    #
    current_error = np.max(np.abs(truth_at_test - cubic_evaluate))
    errors.append(current_error)

    
plt.figure(figsize=(10, 6))
fig, ax = plt.subplots()
plt.loglog(n_values, errors, label='My Error', marker='o', linewidth=2)
#
C = errors[0] * (n_values[0]**2)
ref_line = [C * (float(n)**-4) for n in n_values]
plt.loglog(n_values, ref_line, label='Ref n^-4', linestyle='--')

plt.xlabel('Number of Intervals (n)')
plt.ylabel('Max Error $E(n)$')
plt.title('Error Convergence Analysis (Log-Log Plot)')
plt.grid(True, which="both", ls="-", alpha=0.5)
plt.legend()

# 

plt.show()
```
